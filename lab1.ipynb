```python
#import libaries
import numpy as np
import matplotlib.pylab as plt
import librosa
import librosa.display
```

```python
def plotAudio(filename,samplingRate):
    #open audio file avm.wav with librosa load
    sig , sr = librosa.load(filename,mono=True,sr=samplingRate)
   
    # adjust the X x-axis to represent the time 
    time=np.arange(0,sig.shape[0])/sr
    plt.plot(time,sig)
   
    # scale the y-axis to represent the amplitude
    excursion=max(abs(sig))
    sigNew = sig / excursion
   
    #label the x and y axis by typing in chat
    xlabel = 'Time (seconds)'
    ylabel = 'amplitude'

    plt.xlabel=(xlabel)
    plt.ylabel=(ylabel)
```

```python
plotAudio('avm.wav' ,44100)
```
